```python
# one hot encoding
trf1 = ColumnTransformer([
    ('home_ownership',OneHotEncoder(sparse=False,handle_unknown='ignore'),[6,8])
],remainder='passthrough')
```

```python
#trf2 = GaussianNB()
#trf3 = KNeighborsClassifier()
#trf4 = LogisticRegression()
trf5 = SVC()
#trf6 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([
                ('trf1',trf1),
                #('trf2',trf2)
                # ('trf3',trf3)
                #('trf4',trf4)
                ('trf5',trf5)
                #('trf6',trf6)
    
            ])
```
